## 8) If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? (Profit is defined as interest earned on Monthly Profit. 
#### Monthly Profit = Monthly repayment – Monthly spend. Interest is earned only on positive profits and not on negative amounts)

```python
Customer_Acqusition_spend_Repayment = pd.merge(left=Customer_Acqusition_spend,right=Repayment,on="Customer",how="inner")
Customer_Acqusition_spend_Repayment.head(2)
```

```python
interest_group = Customer_Acqusition_spend_Repayment.groupby(["Year","Month_Name"])['Spend_Amount','Repay_Amount'].sum()
interest_group
```

```python
interest_group['Monthly Profit']=interest_group['Repay_Amount']-interest_group['Spend_Amount']
interest_group
```

```python
# interested earned is 2.9% of Month profit
interest_group['Interest Earned'] = (2.9* interest_group['Monthly Profit'])/100
interest_group
```

## 17) Write user defined PYTHON function to perform the following analysis:
#### You need to find top 10 customers for each city in terms of their repayment amount by different products and by different time periods i.e. year or month. 
#### The user should be able to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the function should automatically take these inputs while identifying the top 10 customers.

```python
Customer_Acqusition_Repayment.head()
```

```python
Customer_Acqusition_Repayment['Month'] = pd.to_datetime(Customer_Acqusition_Repayment['Month'])
Customer_Acqusition_Repayment['Month_Name'] = Customer_Acqusition_Repayment['Month'].dt.month_name()
Customer_Acqusition_Repayment['Year'] = Customer_Acqusition_Repayment['Month'].dt.year
Customer_Acqusition_Repayment.head()
```

```python
def summ(Product, timeperiod):
    print("Enter the product and time period")
    products = ['Gold', 'Platimum', 'Silver']
    periods = ['Month_Name', 'Year']

    if Product in products and timeperiod in periods:
        pivot = Customer_Acqusition_Repayment.pivot_table(index=['Product', 'City', 'Customer'], columns=timeperiod, values='Amount', aggfunc='sum')
        result = pivot.loc[(Product, ['BANGALORE', 'BOMBAY', 'CALCUTTA', 'CHENNAI', 'COCHIN', 'DELHI', 'PATNA', 'TRIVANDRUM']), :]
        return result
    else:
        return "Invalid product or time period."
```
